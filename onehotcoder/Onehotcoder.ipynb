```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('carprices.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
#now doing onehotcoder with the help of pandas function get_dummies() method
df_encoder= pd.get_dummies(df['Car Model'],prefix='Car Model')
df_encoder=df_encoder.astype('int')
df_encoder
```

```python
#now merage the both the data set 
merage = pd.concat([df,df_encoder],axis=1)
merage
```

```python
final_df = merage.drop(['Car Model','Car Model_Mercedez Benz C class'],axis=1)
final_df
```

```python
X = final_df.drop('Sell Price($)',axis = 1)
X
```

```python
y = final_df['Sell Price($)']
y
```

```python
#now making the model using the linear Regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[45000,4,1,0]])
```

```python
model.predict([[69000,6,0,1]])
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.score(X,y)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
#ploting the year vs per capital income
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'],df['per capita income (US$)'],color='red',marker='*')
```

```python
df_new = df.drop('per capita income (US$)',axis=1)
df_new
```

```python
per_captial_income = df['per capita income (US$)']
per_captial_income.head()
```

```python
#now wmaking the model 
reg = linear_model.LinearRegression()
reg.fit(df_new,per_captial_income)
```

```python
#prediction 
reg.predict([[1970]])
```

```python
reg.predict([[2015]])
```

```python
reg.predict([[2016]])
```

```python
reg.predict([[2021]])
```
